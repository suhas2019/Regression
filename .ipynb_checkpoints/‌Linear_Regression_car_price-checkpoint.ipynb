### <font color='blue'> Step5a - Building linear regression model using statsmodels package:

### <font color='blue'> We will first build a model with manual feature elimination </font>

#### First lets create X (independent) and y (dependent) variables

```python
y_train = df_train.pop('price')
X_train = df_train.copy()
```

#### Now lets start building a model considering all independent variables and try eleminating the ones which are insignificant in predicting price. Manual feature elimination.

#### We will use statsmodel package as it helps us by providing statistical significance of the model

```python
# Adding a constant to X_train to obtain the y-intercept from statsmodel

X_train_sm = sm.add_constant(X_train)

# create an instance of statsmodels OLS using y_train and X_train_sm and then fit
lr_sm_obj = sm.OLS(y_train, X_train_sm)
lr_sm = lr_sm_obj.fit()
```

```python
print(lr_sm.summary())
```

### Observations from the summary of the model:
#### This model has a R-squared of 0.976 and adjusted R-squared of 0.957.
#### some of the co-efficents look insignificant compared to a significance level of 5%.

#### Now lets look at the VIF (Variance inflation Factor) of features for comparision with p-value from summary.

```python
vif = pd.DataFrame()
vif['Feature'] = X_train.columns.tolist()
vif['VIF'] = [round(variance_inflation_factor(X_train.values, i),2) for i in range(X_train.shape[1])]
vif.sort_values(by='VIF', ascending=False, inplace=True)
vif
```

#### VIF of some of the columns is inf meaning infinte. This tells that, all values of these can be completely described by other column values.

#### Considering a significance level of 5%, we can say that any co-efficient with a p-value > 0.05 is insignificant and can be eliminated from the model.
#### So we can continue this process of manual feature elimination until all the co-efficients in the model are significant (i.e. p-value < 0.05).

```python
# setting the significance level at 5%.
significance_level = 0.05

#max_pvalue from previous model.
df_result = lr_sm.summary2().tables[1]
df_result.drop('const', axis=0, inplace=True)
max_pvalue = max(df_result['P>|t|'])

# Loop to perform feature elimination
while (max_pvalue > significance_level):
    
    #Get the feature with max_pvalue
    eliminating_feature = df_result.index[df_result['P>|t|'] == max_pvalue].tolist()[0]
    
    #print the feature to be eliminated and it p-value
    print('Eliminating Feature: ' + eliminating_feature + ' | p-value : ', max_pvalue)
    
    #remove this feature from the model (X_train)
    X_train.drop(columns=eliminating_feature, inplace=True)
    
    # Adding a constant to X_train to obtain y-intercept from statsmodel
    X_train_sm = sm.add_constant(X_train)

    # create an instance of statsmodels OLS using y_train and X_train_sm and then fit
    lr_sm = sm.OLS(y_train, X_train_sm).fit()
    
    #Get the summary of model into a dataframe, tables is a list and second element contains coefficeints and pvalues.
    df_result = lr_sm.summary2().tables[1]
    
    #Remove the p-value for y-intercept before selecting feature with maximum p-value
    df_result.drop('const', axis=0, inplace=True)
    
    #Calculate max_pvalue
    max_pvalue = max(df_result['P>|t|'])
#
```

#### We will continue this process of manual feature elimination using VIF with a maximum permissible value of 5.00

```python
#check VIF after above feature elimination
vif = pd.DataFrame()
vif['Feature'] = X_train.columns.tolist()
vif['VIF'] = [round(variance_inflation_factor(X_train.values, i),2) for i in range(X_train.shape[1])]
vif.sort_values(by='VIF', ascending=False, inplace=True)
max_vif = vif.iloc[0][1]

# perform elimination until all features in the model have VIF below 5.00
while (max_vif > 5.00):
    
    #Get the feature with maximum VIF
    eliminating_feature = vif.iloc[0][0]
    
    #print the eliminating feature which is having maximum VIF value
    print('Eliminating Feature: ' + eliminating_feature + ' | VIF : ', max_vif)
    
    #remove this feature from the model (X_train)
    X_train.drop(columns=eliminating_feature, inplace=True)
    
    # Adding a constant to X_train to obtain y-intercept from statsmodel
    X_train_sm = sm.add_constant(X_train)

    # create an instance of statsmodels OLS using y_train and X_train_sm and then fit
    lr_sm = sm.OLS(y_train, X_train_sm).fit()
    
    #Get the summary of model into a dataframe, tables is a list and second element contains coefficeints and pvalues.
    df_result = lr_sm.summary2().tables[1]
    
    #Remove the p-value for y-intercept before selecting feature with maximum p-value
    df_result.drop('const', axis=0, inplace=True)
    
    #Calculate max_pvalue
    max_pvalue = max(df_result['P>|t|'])
    #
    while (max_pvalue > significance_level):
    
        #Get the feature with max_pvalue
        eliminating_feature = df_result.index[df_result['P>|t|'] == max_pvalue].tolist()[0]

        #print the feature to be eliminated and it p-value
        print('Eliminating Feature: ' + eliminating_feature + ' | p-value : ', max_pvalue)
        
        #remove this feature from the model (X_train)
        X_train.drop(columns=eliminating_feature, inplace=True)

        # Adding a constant to X_train to obtain y-intercept from statsmodel
        X_train_sm = sm.add_constant(X_train)

        # create an instance of statsmodels OLS using y_train and X_train_sm and then fit
        lr_sm = sm.OLS(y_train, X_train_sm).fit()

        #Get the summary of model into a dataframe, tables is a list and second element contains coefficeints and pvalues.
        df_result = lr_sm.summary2().tables[1]

        #Remove the p-value for y-intercept before selecting feature with maximum p-value
        df_result.drop('const', axis=0, inplace=True)

        #Calculate max_pvalue
        max_pvalue = max(df_result['P>|t|'])
    #
    #check VIF after recursive elimination
    vif = pd.DataFrame()
    vif['Feature'] = X_train.columns.tolist()
    vif['VIF'] = [round(variance_inflation_factor(X_train.values, i),2) for i in range(X_train.shape[1])]
    vif.sort_values(by='VIF', ascending=False, inplace=True)
    #calculate maximum VIF again
    max_vif = vif.iloc[0][1]
```

#### Final Model and its summary using statsmodels package and manual feature elimination:

```python
#print the summary of model
print()
print('__________________________________________________________________________')
print('      MODEL USING statsmodels PACKAGE AND MANUAL FEATURE ELIMINATION      ')
print('__________________________________________________________________________')
print()
print(lr_sm.summary())
```

### <font color='blue'> Step5b - Building linear regression model using sklearn package:

### <font color='blue'> We will use Recursive Feature Elimination (RFE) for selecting best features. </font>

```python
#Lets check df_train
df_train.shape
```

```python
# We have the y_train and y_test created above and can be used as is.
# But we have to create X_train and X_test again for RFE and model building using scikit learn
X_train = df_train.copy()
X_test = df_test.copy()
```

```python
#Lets start by building the linear regression model using sklearn

#Create an instance of LinerRegression
lr_sk_obj = LinearRegression()

#fit the model using y_train and X_train
lr_sk = lr_sk_obj.fit(X_train, y_train)
```

#### Let's start by selecting 15 best features using RFE

```python
#Using sklearn RFE for selecting the best 8 predictors. Create an RFE instance and fit.
sk_rfe = RFE(lr_sk, 15)
sk_rfe = sk_rfe.fit(X_train, y_train)
```

```python
#Looking at the selected and rejected columns by RFE

print('Columns Selected are:')
print('_____________________')
print(X_train.columns[sk_rfe.support_])
print()
print('Columns Rejected are:')
print('_____________________')
print(X_train.columns[~sk_rfe.support_])
```

#### Using these 15 features given by RFE, we can build model using statsmodels so that we can look at the statistics summary of the model.

```python
# Drop columns rejected by RFE from X_train
X_train.drop(columns=X_train.columns[~sk_rfe.support_].tolist(),inplace=True)
X_train_sm = sm.add_constant(X_train)
lr_sm2 = sm.OLS(y_train, X_train_sm).fit()

#Lets look at the summary of this model
print(lr_sm2.summary())

#Lets look at the VIF of the features given by RFE
vif = pd.DataFrame()
vif['Feature'] = X_train.columns.tolist()
vif['VIF'] = [round(variance_inflation_factor(X_train.values, i),2) for i in range(X_train.shape[1])]
vif.sort_values(by='VIF', ascending=False, inplace=True)
vif
```

#### This model has an R-squared of 0.939 and Adjusted R-squared as 0.933
#### The p-values of all coefficients look good, but the VIF values are high for some features.
#### We can iterate through VIF and p-values to remove insignificant features and arrive at final model

```python
# Get maximum value of VIF
max_vif = vif.iloc[0][1]

# perform elimination until all features in the model have VIF below 5.00 and p-values below 0.05
while (max_vif > 5.00):
    
    #Get the feature with maximum VIF
    eliminating_feature = vif.iloc[0][0]
    
    #print the eliminating feature which is having maximum VIF value
    print('Eliminating Feature: ' + eliminating_feature + ' | VIF : ', max_vif)
    
    #remove this feature from the model (X_train)
    X_train.drop(columns=eliminating_feature, inplace=True)
    
    # Adding a constant to X_train to obtain y-intercept from statsmodel
    X_train_sm = sm.add_constant(X_train)

    # create an instance of statsmodels OLS using y_train and X_train_sm and then fit
    lr_sm2 = sm.OLS(y_train, X_train_sm).fit()
    
    #Get the summary of model into a dataframe, tables is a list and second element contains coefficeints and pvalues.
    df_result = lr_sm2.summary2().tables[1]
    
    #Remove the p-value for y-intercept before selecting feature with maximum p-value
    df_result.drop('const', axis=0, inplace=True)
    
    #Calculate max_pvalue
    max_pvalue = max(df_result['P>|t|'])
    #
    while (max_pvalue > significance_level):
    
        #Get the feature with max_pvalue
        eliminating_feature = df_result.index[df_result['P>|t|'] == max_pvalue].tolist()[0]

        #print the feature to be eliminated and it p-value
        print('Eliminating Feature: ' + eliminating_feature + ' | p-value : ', max_pvalue)
        
        #remove this feature from the model (X_train)
        X_train.drop(columns=eliminating_feature, inplace=True)

        # Adding a constant to X_train to obtain y-intercept from statsmodel
        X_train_sm = sm.add_constant(X_train)

        # create an instance of statsmodels OLS using y_train and X_train_sm and then fit
        lr_sm2 = sm.OLS(y_train, X_train_sm).fit()

        #Get the summary of model into a dataframe, tables is a list and second element contains coefficeints and pvalues.
        df_result = lr_sm2.summary2().tables[1]

        #Remove the p-value for y-intercept before selecting feature with maximum p-value
        df_result.drop('const', axis=0, inplace=True)

        #Calculate max_pvalue
        max_pvalue = max(df_result['P>|t|'])
    #
    #check VIF after recursive elimination
    vif = pd.DataFrame()
    vif['Feature'] = X_train.columns.tolist()
    vif['VIF'] = [round(variance_inflation_factor(X_train.values, i),2) for i in range(X_train.shape[1])]
    vif.sort_values(by='VIF', ascending=False, inplace=True)
    #calculate maximum VIF again
    max_vif = vif.iloc[0][1]
```

```python
#Lets look at the summary of this model and VIF values
print(lr_sm2.summary())
vif
```

#### This model has R-Squared of 0.883, adjusted R-Squared of 0.878 and prob (F-statistic) of 8.77e-61 (zero).
#### But we see that feature 'cylindernumber_twelve' has a p-value of 0.027. So in order to reduce the number of features and make model light, we can try removing this feature from the model if the impact on R-Squared is almost nill.

```python
# drop feature 'cylindernumber_twelve'
X_train.drop(columns=['cylindernumber_twelve'],inplace=True)
X_train_sm = sm.add_constant(X_train)
lr_sm2 = sm.OLS(y_train, X_train_sm).fit()

#Lets look at the summary of this model
print()
print('__________________________________________________________________________')
print('           MODEL USING RFE FROM sklearn AND statsmodels PACKAGE           ')
print('__________________________________________________________________________')
print()
print(lr_sm2.summary())

#Lets look at the VIF of the features given by RFE from sklearn
vif = pd.DataFrame()
vif['Feature'] = X_train.columns.tolist()
vif['VIF'] = [round(variance_inflation_factor(X_train.values, i),2) for i in range(X_train.shape[1])]
vif.sort_values(by='VIF', ascending=False, inplace=True)
vif
```
