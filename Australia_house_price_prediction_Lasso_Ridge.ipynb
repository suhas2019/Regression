### Problem Statement:
### -----------------------------

### A US-based housing company named Surprise Housing has decided to enter the Australian market. The company uses data analytics to purchase houses at a price below their actual values and flip them on at a higher price. ### <font color = 'blue'> Step3: Preprocessing - Dummy variable creation, train test split and scaling </font>

```python
#Lets start by taking the list of numerical and categorical columns
numcols = dfhousing.select_dtypes(include=[np.number]).columns.tolist()
catcols = dfhousing.select_dtypes(exclude=[np.number]).columns.tolist()
```

```python
dfhousing_catg = dfhousing.loc[:,catcols]
#
dfhousing_catg = pd.get_dummies(data=dfhousing_catg, drop_first=True)
dfhousing = pd.concat([dfhousing, dfhousing_catg], axis=1)
dfhousing.drop(columns=catcols, inplace=True)
dfhousing.head(5)
```

```python
len(dfhousing.columns)
```

```python
#Splitting the data into train and test datasets

df_train, df_test = train_test_split(dfhousing, train_size=0.7, test_size=0.3, random_state=100)
print('Length of train set : ', len(df_train))
print('Length of test set : ', len(df_test))
```

```python
#Scaling all the numerical columns using StandardScaler
std_scaler = StandardScaler()
df_train[numcols] = std_scaler.fit_transform(df_train[numcols])
#
df_test[numcols] = std_scaler.transform(df_test[numcols])
```

```python
#Seperate the target as y and rest as X feature vectors
X_train = df_train.copy()
y_train = X_train.pop('SalePrice')
#
X_test = df_test.copy()
y_test = X_test.pop('SalePrice')
```

```python
#Lets calculate the correlation coefficient between variables
Xcorr = X_train.corr().abs()

# Get upper triangle of correlation matrix
upper = Xcorr.where(np.triu(np.ones(Xcorr.shape), k=1).astype(np.bool))

# Get all columns which display more than 80% correlation
high_corr_cols = [col for col in upper.columns if any(upper[col] > 0.80)]
high_corr_cols
```

### Lets drop these highly correlated columns

```python
X_train.drop(columns=high_corr_cols, inplace=True)
X_test.drop(columns=high_corr_cols, inplace=True)
```

### <font color='blue'> Step4a: Model Building - Lasso </font>

```python
#Lets do a KFold cross validation for tuning the hyperparameter alpha for lasso using GridSearchCV

folds = KFold(n_splits=5, shuffle=True, random_state=100)
#
hyper_params = [{'alpha':[0.0001, 0.00025, 0.0005, 0.00075, 0.001, 0.0025, 0.005, 0.0075, 0.01, 0.05, 0.1]}]

#create a lasso object
lasso = Lasso()

#create GridSearchCV object
cv_lasso = GridSearchCV(estimator=lasso,
                        param_grid=hyper_params,
                        scoring='r2',
                        cv=folds,
                        return_train_score=True,
                        verbose=1)
#
cv_lasso.fit(X_train, y_train)
```

```python
#Looking at the cross validation results.

cv_lasso_results = pd.DataFrame(cv_lasso.cv_results_)
cv_lasso_results.head(5)
```

```python
print('Best r2 score : ', cv_lasso.best_score_)
print('Best hyperparameter value : ', cv_lasso.best_params_)
```

```python
#Lets look at the plot of train versus test r2_score as given cross validation

plt.figure(figsize=(10,5))
plt.plot(cv_lasso_results['param_alpha'], cv_lasso_results["mean_test_score"])
plt.plot(cv_lasso_results['param_alpha'], cv_lasso_results["mean_train_score"])
plt.xlabel('Hyperparameter alpha')
plt.ylabel('r-squared')
plt.title('Lasso Cross Validation Results')
plt.legend(['test score', 'train score'], loc='upper right')
```

### The above plot of train versus test r2score from cross validation shows the below:
### 1. r2score for train starts with a high value and decreases further
### 2. r2score for test increases initially and decreases later
### 3. The gap between train and test r2score narrows down from alpha = 0.0005
### 4. Considering this gap and the r-squared value the optimal value for alpha is in the range 0.005 to 0.02

```python
#Lets build the lasso model with alpha=0.01 and 
#look at the non zero coefficients and
#evalute on test set

lasso_tuned = Lasso(alpha=0.01).fit(X_train, y_train)
print('Number of non-zero Coefficients {}'.format(np.sum(lasso_tuned.coef_!=0)))
print()
y_train_pred = lasso_tuned.predict(X_train)
print('Train r-squared : ', r2_score(y_true=y_train, y_pred=y_train_pred))
print()
y_test_pred = lasso_tuned.predict(X_test)
print('Test r-squared : ', r2_score(y_true=y_test, y_pred=y_test_pred))
```

```python
#Lets look at the top 10 coefficeints from lasso
df_lasso_coef = pd.DataFrame({'Features':X_train.columns.tolist(),'Coefficients':list(lasso_tuned.coef_), 'Absvalue': list(abs(lasso_tuned.coef_))})
df_lasso_coef.sort_values(by=['Absvalue'], ascending=False, inplace=True)
df_lasso_coef.head(15)
```

### <font color='blue'> Step4b: Model Building - Ridge </font>

```python
#Lets do a KFold cross validation for tuning the hyperparameter alpha for ridge using GridSearchCV

folds = KFold(n_splits=5, shuffle=True, random_state=100)
#
hyper_params = [{'alpha':[0.1, 1.0, 10, 50, 100, 250, 500, 1000]}]

#create a ridge object
ridge = Ridge()

#create GridSearchCV object
cv_ridge = GridSearchCV(estimator=ridge,
                        param_grid=hyper_params,
                        scoring='r2',
                        cv=folds,
                        return_train_score=True,
                        verbose=1)
#
cv_ridge.fit(X_train, y_train)
```

```python
#Looking at the cross validation results.

cv_ridge_results = pd.DataFrame(cv_ridge.cv_results_)
cv_ridge_results.head(5)
```

```python
print('Best r2 score : ', cv_ridge.best_score_)
print('Best hyperparameter value : ', cv_ridge.best_params_)
```

```python
#Lets look at the plot of train versus test r2_score as given cross validation

plt.figure(figsize=(10,5))
plt.plot(cv_ridge_results['param_alpha'], cv_ridge_results["mean_test_score"])
plt.plot(cv_ridge_results['param_alpha'], cv_ridge_results["mean_train_score"])
plt.xlabel('Hyperparameter alpha')
plt.ylabel('r-squared')
plt.title('Ridge Cross Validation Results')
plt.legend(['test score', 'train score'], loc='upper right')
```

### The above plot of train versus test r2score from cross validation shows the below:
### 1. r2score for train starts with a high value and decreases further
### 2. r2score for test increases initially and decreases later
### 3. The gap between train and test r2score narrows down from alpha = 10

#### Question 4:
#### How can you make sure that a model is robust and generalisable? What are the implications of the same for the accuracy of the model and why?

#### Answer:

#### Regularization is the technique for making the model robust and generalisable. Such a model will exhibit less variance but will be more bias. Linear regression learning algorithm will have a component for computing least squared error and also a component for regularizing the model. This regularization component is the sum of squared coefficients in ridge and sum of absolute value of coefficients in lasso.

#### The accuracy will be less in a regularized model compared to unregularized model. This is because regularization reduces the value of coefficients in order to stop the model from becoming more complex. This makes the model less accurate in terms of predicting the target, but will become more robust and generalisable.

#### The hyperparameter lambda is a scalar which decides the weight of regularization component in the learning algorithm. More is the value of lambda, more is the regularization done by the model. More is the regularization, less will be the model complexity.
